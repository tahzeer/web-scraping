```python
import re
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://en.wikipedia.org/wiki/List_of_oldest_companies"
```

```python
page = requests.get(url)
data = pd.read_html(page.content)
```

```python
df = pd.DataFrame()
```

```python
for d in data:
    df = pd.concat([df, d])

df
```

```python
df.info()
```

```python
df.drop(columns=['Sources'], inplace=True)
```

```python
df.info()
```

```python
df = df.reset_index().drop(columns=['index'])
```

```python
df
```

```python
df['Current location'].value_counts()
```

```python
for year in df['Year']:
    if year.isdigit():
        continue
    else:
        print(year + "  " + re.search(r'[0-9]{4}', year).group() + "\n")
        # df['Year'].replace(year, re.search(r'[0-9]{4}', year).group())
```

```python
for year in df['Year']:
    if year.isdigit():
        continue
    else:
        df['Year'].replace(year, re.search(r'[0-9]{4}', year).group(), inplace=True)
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.to_csv('oldest_companies.csv',index=False, encoding='utf-8')
```
